```python
#15Probability density distribution of continuous
#numerical variable-length
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data and replace '.' with NaN, then convert to numeric
data = pd.read_csv('Car_sales.csv')
data['Length'] = pd.to_numeric(data['Length'], errors='coerce')

# Plot the KDE for Length
sns.kdeplot(data=data, x='Length')
plt.title('Probability Density of Length')
plt.show()
```

```python
#18 Print correlation coefficient value of price and sales
import pandas as pd

# Load the data and convert columns to numeric, replacing '.' with NaN
data = pd.read_csv('Car_sales.csv')
data['Price in thousands'] = pd.to_numeric(data['Price in thousands'], errors='coerce')
data['Sales in thousands'] = pd.to_numeric(data['Sales in thousands'], errors='coerce')

# Calculate correlation
correlation = data['Price in thousands'].corr(data['Sales in thousands'])
print(f"Correlation between Price and Sales: {correlation}")
```

```python
#22 Boxplot of other numerical variables w.r.t manufacture
import pandas as pd
import matplotlib.pyplot as plt

# Load the data and convert numeric columns to handle non-numeric values
data = pd.read_csv('Car_sales.csv')
numerical_cols = ['Price in thousands', 'Horsepower', 'Length', 'Fuel efficiency']
for col in numerical_cols:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Create boxplots
for col in numerical_cols:
    data.boxplot(column=col, by='Manufacturer')
    plt.xticks(rotation=90)
    plt.show()
```

```python
#24 Encode other categorical variables using label encoder
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv('Car_sales.csv')
le = LabelEncoder()
categorical_cols = ['Manufacturer', 'Model', 'Vehicle type']
for col in categorical_cols:
    data[col] = le.fit_transform(data[col])
print(data.head())
```

```python
#25 Encode categorical variable vehicle type using one-hot
encoder
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

data = pd.read_csv('Car_sales.csv')
one_hot = OneHotEncoder(sparse_output=False)  # Updated from sparse=False to sparse_output=False
vehicle_type_encoded = one_hot.fit_transform(data[['Vehicle type']])
vehicle_type_df = pd.DataFrame(vehicle_type_encoded, columns=one_hot.get_feature_names_out(['Vehicle type']))
data = pd.concat([data.drop('Vehicle type', axis=1), vehicle_type_df], axis=1)
print(data.head())
```

```python
#26 Split the data set into train and test set 70% train set
#10% test set
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv('Car_sales.csv')
X = data.drop('Sales in thousands', axis=1)
y = data['Sales in thousands']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
print(f"Train set size: {len(X_train)}, Test set size: {len(X_test)}")
```
