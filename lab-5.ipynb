## Handling Missing Values

```python
#Replace the msiing values in Embarked column by 'S'
import pandas as pd

# Step 1: Load the dataset into a DataFrame
# Assuming the dataset is in a CSV file named 'train.csv'
train = pd.read_csv('train.csv')

# Step 2: Replace missing values in the Embarked column with 'S'
train['Embarked'].fillna('S', inplace=True)

# Step 3: Verify if missing values are replaced
print(train['Embarked'].isnull().sum())
```

```python
#Impute the missing values in Fare column with median value
# Impute missing values in Fare column with median value
train = pd.read_csv('train.csv')
train['Fare'].fillna(train['Fare'].median(), inplace=True)

# Verify if missing values are replaced
print(train['Fare'].isnull().sum())
```

```python
import numpy as np

# Calculate mean and standard deviation of non-missing Age data
train = pd.read_csv('train.csv')
mean_age = train['Age'].mean()
std_age = train['Age'].std()

# Fill missing Age values with random numbers drawn from a normal distribution
# with the same mean and standard deviation as the non-missing Age data
train['Age'].fillna(pd.Series(np.random.normal(mean_age, std_age, size=len(train))), inplace=True)

# Verify if missing values are replaced
print(train['Age'].isnull().sum())
```

```python
# Create new feature IsAlone from FamilySize
#If FamilySize == 1 then IsAlone = True else IsAlone = False
# Create new feature IsAlone from FamilySize
train['IsAlone'] = train['FamilySize'].apply(lambda x: True if x == 1 else False)

# Display the first few rows to verify the new feature
print(train[['FamilySize', 'IsAlone']].head())
```
